Retrieving NYPD Stop and Frisk Data through Webscraping:

I used the below function to check if the webpage is accesible. I then used a for loop to return the content based on certain conditions:

```python
url = 'https://www.nyclu.org/en/stop-and-frisk-data' 

# Years to extract data for
years_to_extract = ['2019','2020', '2021', '2022']

# I will get the request fot the HTML.
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # I will find the list items within <ul>
    list_items = soup.find_all('li')

    # I will then store a dictionary to create a for loop to extract information.
    data_per_year = {}

    # Loop through each list item to extract information for each year,
    for item in list_items:
        strong_tag = item.find('strong')
        p_tag = item.find('p')

        if strong_tag and p_tag:
            year = strong_tag.text
            if year in years_to_extract:
                data = p_tag.text
                data_per_year[year] = data

    # To display data for the specified years:
    for year in years_to_extract:
        year_data = data_per_year.get(year, f'Data for {year} not found')
        print(f"Data for {year}:")
        print(year_data)
else:
    print("Failed to fetch the webpage")
```

3. Save NYPD Stop and Frisk to Dataframe and conduct checks:

```python

text_data = {
    'Data for 2019': 'Data for 2019 not found',
    'Data for 2020': 'In 2020, 9,544 stops were recorded.5,791 were innocent (61 percent).5,367 were Black (56 percent).2,879 were Latinx (30 percent).856 were white (9 percent).219 were Asian / Pacific Islander (2 percent)132 were Middle Eastern/Southwest Asian (1.4 percent)',
    'Data for 2021': 'In 2021, 8,947 stops were recorded.5,422 were innocent (61 percent).5,404 were Black (60 percent).2,457 were Latinx (27 percent).732 were white (8 percent).192 were Asian / Pacific Islander (2 percent)71 were Middle Eastern/Southwest Asian (1 percent)',
    'Data for 2022': 'In 2022, 15,102 stops were recorded.9,750 were not arrested or given a summons (65 percent).8,863 were Black (59 percent). 4,477 were Latinx (30 percent).1,077 were white (7 percent).320 were Asian / Pacific Islander (2 percent).148 were Middle Eastern/Southwest Asian (1 percent).'
}
#I will create empty dictionaries to store the extracted data
data = {
    'Year': [],
    'Stops': [],
    'Innocent': [],
    'Black': [],
    'Latinx': [],
    'White': [],
    'Asian/Pacific Islander': [],
    'Middle Eastern/Southwest Asian': []
}
# Iterate through the text data
for year, info in text_data.items():
    data['Year'].append(year.split()[-1])
    
# Conducting checks to see if the information for the year is not available

# Add zeros for each category and continue to the next year
    if 'not found' in info:
        data['Stops'].append(0)
        data['Innocent'].append(0)
        data['Black'].append(0)
        data['Latinx'].append(0)
        data['White'].append(0)
        data['Asian/Pacific Islander'].append(0)
        data['Middle Eastern/Southwest Asian'].append(0)
        continue
    # Extract the number of stops for the year
    stops = re.search(r'(\d[\d,]*) stops', info)
    data['Stops'].append(int(stops.group(1).replace(',', '')) if stops else 0)
    
    # Extract the number of innocent individuals
    innocent = re.search(r'(\d[\d,]*) were innocent', info)
    data['Innocent'].append(int(innocent.group(1).replace(',', '')) if innocent else 0)
    
    #Extract rest of the relevant data.
    black = re.search(r'(\d[\d,]*) were Black', info)
    data['Black'].append(int(black.group(1).replace(',', '')) if black else 0)
    
    latinx = re.search(r'(\d[\d,]*) were Latinx', info)
    data['Latinx'].append(int(latinx.group(1).replace(',', '')) if latinx else 0)
    
    white = re.search(r'(\d[\d,]*) were white', info)
    data['White'].append(int(white.group(1).replace(',', '')) if white else 0)
    
    asian_pacific = re.search(r'(\d[\d,]*) were Asian / Pacific Islander', info)
    data['Asian/Pacific Islander'].append(int(asian_pacific.group(1).replace(',', '')) if asian_pacific else 0)
    
    middle_eastern = re.search(r'(\d[\d,]*) were Middle Eastern/Southwest Asian', info)
    data['Middle Eastern/Southwest Asian'].append(int(middle_eastern.group(1).replace(',', '')) if middle_eastern else 0)

#I will now create the dataframe
stop_and_frisk = pd.DataFrame(data)
stop_and_frisk
```

I would like to implement a test-driven development (TDD) approach to ensure that the above code runs as expected:

```python
#Defining a filtering function

def filter_crime_by_month_year(data, year, month):
    """
    Filter crime data based on the specified year and month.

    Parameters:
    - data: DataFrame, the crime data
    - year: int, the year for filtering
    - month: int, the month for filtering

    Returns:
    - DataFrame, filtered crime data
    """
    filtered_data = data[(data['Year'] == year) & (data['Month'] == month)]
    return filtered_data
```

```python
#Write test cases for the filtering function

class TestCrimeFiltering(unittest.TestCase):
    def setUp(self):
        # Sample crime data for testing
        date_values = ['2021-01-15', '2021-02-20', '2021-02-25', '2022-03-10']
        self.crime_data = pd.DataFrame({'Date of Arrest': date_values})
        self.crime_data['Date of Arrest'] = pd.to_datetime(self.crime_data['Date of Arrest'])
        self.crime_data['Year'] = self.crime_data['Date of Arrest'].dt.year
        self.crime_data['Month'] = self.crime_data['Date of Arrest'].dt.month

    def test_filter_by_month_year(self):
        # Test case 1: Check if filtering works for a specific year and month
        result_data = filter_crime_by_month_year(self.crime_data, 2021, 2)
        expected_data = self.crime_data[(self.crime_data['Year'] == 2021) & (self.crime_data['Month'] == 2)]
        pd.testing.assert_frame_equal(result_data, expected_data)

        # Test case 2: Check if filtering works for a different year and month
        result_data = filter_crime_by_month_year(self.crime_data, 2022, 3)
        expected_data = self.crime_data[(self.crime_data['Year'] == 2022) & (self.crime_data['Month'] == 3)]
        pd.testing.assert_frame_equal(result_data, expected_data)

        # Add more test cases based on the behavior of your filtering function

# Create a test suite
test_suite = unittest.TestLoader().loadTestsFromTestCase(TestCrimeFiltering)

# Run the tests
unittest.TextTestRunner().run(test_suite)
```

I created a simple function to perform the filtering operation based on the provided month and year. The results show that the test has been passed and everything runs correctly.

I created a condition to slice my data and isolate it to analyse the information for each year seperately. I specifically picked these years as I was intesrested in exploring the crime patterns before, during and after the mass protests of 2020.

More specifically I want to further slice my data to seperate it into pre-protest; during protest and post periods. I will also create a column called 'Temporal Period' based on protest dates as this will be useful for further analysis:

```python
# Step 1: Create a new column 'Temporal Period' based on protest dates
filtered_crime_19_20_21_22 = filtered_crime_19_20_21_22.copy()
filtered_crime_19_20_21_22['Temporal Period'] = pd.cut(
    pd.to_datetime(filtered_crime_19_20_21_22['Date of Arrest']), 
    bins=[pd.to_datetime('2006-01-01'), pd.to_datetime('2020-05-31'), pd.to_datetime('2020-12-31'), pd.to_datetime('2022-12-31')],
    labels=['Pre-Protest', 'During Protest', 'Post-Protest']
)

# preceding_crime: Up to May 2020
pre_protest_condition = (
    (filtered_crime_19_20_21_22['Year'] >= 2006) & (filtered_crime_19_20_21_22['Year'] < 2020) |
    ((filtered_crime_19_20_21_22['Year'] == 2020) & (filtered_crime_19_20_21_22['Month'] <= 5))
)
pre_protest = filtered_crime_19_20_21_22.loc[pre_protest_condition].copy()
pre_protest['Temporal Period'] = 'Pre-Protest'

# during_crime: June 2020 to December 2020
during_protest_condition = (
    ((filtered_crime_19_20_21_22['Year'] == 2020) & (filtered_crime_19_20_21_22['Month'] >= 6)) |
    ((filtered_crime_19_20_21_22['Year'] == 2020) & (filtered_crime_19_20_21_22['Month'] <= 12))
)
during_protest = filtered_crime_19_20_21_22.loc[during_protest_condition].copy()
during_protest['Temporal Period'] = 'During Protest'

# after_crime: January 2021 to December 2022
post_protest_condition = (
    ((filtered_crime_19_20_21_22['Year'] == 2021) & (filtered_crime_19_20_21_22['Month'] >= 1)) |
    ((filtered_crime_19_20_21_22['Year'] > 2021) & (filtered_crime_19_20_21_22['Year'] < 2022)) |
    ((filtered_crime_19_20_21_22['Year'] == 2022) & (filtered_crime_19_20_21_22['Month'] <= 12))
)
post_protest = filtered_crime_19_20_21_22.loc[post_protest_condition].copy()
post_protest['Temporal Period'] = 'Post-Protest'

#previewing the data

#pre_protest
#during_protest
#post_protest
```

### 7.1 Clustering Algorithm: K-Modes

I will first use K-Modes clustering to identify patterns and clusters between different temporal periods. Since, the data is categorical, K-Modes may be more suitable than K-means as it aims to group similar rows based on their most frequent values of categorical features. Unlike, K-Means that is designed for numerical data. 

```python
#Importing the necessary library
from sklearn.preprocessing import LabelEncoder

# Setting a random seed for the reproducibility of results
np.random.seed(42)

#Preparing Data for Clustering by selecting relevant columns
clustering_data = combined_data[['Perpetrator Race', 'Temporal Period']]

#Encoding Categorical Data for Perpatrator Race and Temporal Period:
le = LabelEncoder()
clustering_data_encoded = clustering_data.apply(le.fit_transform)

#2. Determining the Optimal K (Number of Clusters) using elbow method on silhouette score
from kmodes.kmodes import KModes
from sklearn.metrics import silhouette_score

# Trying different values of K
k_values = range(2, 6)
silhouette_scores = []

for k in k_values:
    km = KModes(n_clusters=k, init='Huang', n_init=5, verbose=0)
    clusters = km.fit_predict(clustering_data_encoded)
    silhouette_scores.append(silhouette_score(clustering_data_encoded, clusters))

# Plotting the silhouette scores for different K values
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal K')
plt.show()

#3. Performing K-Modes Clustering:
optimal_k = 3  # Replace with the chosen K value
km = KModes(n_clusters=optimal_k, init='Huang', n_init=5, verbose=0)
clusters = km.fit_predict(clustering_data_encoded)

# Adding the cluster labels to the original data
combined_data['Cluster'] = clusters

#4.Analysing results the clusters to check if there's a noticeable difference in the distribution of black individuals across temporal periods within each cluster.

cluster_analysis = combined_data.groupby(['Cluster', 'Temporal Period', 'Perpetrator Race']).size().reset_index(name='Counts')
print(cluster_analysis)
```

From these results I can see that cluster 0 has a higher count of black individuals during the protest period when compared to other races. Cluster 1 seems to have a relatively balanced distribution across races during the protest and post-protest periods. It also has a significant count of black individuals post-protest. Cluster 2 also has a higher count of black individuals post-protest when compared to other races.


Now I will analyse the distribution of black individuals within each cluster across the different temporal periods. I will first calculate the proportion of black individuals within each cluster for each period and then visualise this.

```python
# Calculate the proportion of black individuals within each cluster for each temporal period
cluster_proportions = cluster_analysis.pivot_table(index=['Cluster', 'Temporal Period'], columns='Perpetrator Race', values='Counts', fill_value=0)
cluster_proportions['Proportion Black'] = cluster_proportions['BLACK'] / cluster_proportions.sum(axis=1)

# Visualise the proportion of black individuals within each cluster for each temporal period
plt.figure(figsize=(12, 8))
sns.barplot(x='Cluster', y='Proportion Black', hue='Temporal Period', data=cluster_proportions.reset_index())
plt.title('Proportion of Black Individuals in Clusters Across Temporal Periods')
plt.xlabel('Cluster')
plt.ylabel('Proportion of Black Individuals')
plt.show()
```

### 7.2 Hypothesis Testing

#### Hypothesis: ANOVA Test

To test the hypothesis of the significant difference in the distribution of black individuals pre-protest, during and post protest within the clusters I will perform a one-way ANOVA test to evaluate whether the means of the distributions are significantly different.

```python
from scipy.stats import f_oneway

# Perform ANOVA on the proportions of black individuals within each cluster for each temporal period
non_empty_clusters = cluster_proportions.index.get_level_values('Cluster').unique()

# Perform ANOVA on proportions of black individuals within each cluster for each temporal period
clusters_data = [
    cluster_proportions.loc[cluster, 'Proportion Black'].values
    for cluster in non_empty_clusters
]

f_stat, p_value = f_oneway(*clusters_data)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

```

The F-statistic and p-value obtained from the ANOVA test suggests that there is a statistically significant difference in the proportions of black individuals within the clusters across temporal periods. The small p-value (0.0154) indicates that the null hypothesis can be rejected, suggesting that at least one group's proportion of black individuals is different from the others. Therefore, this implies that this implies that there is evidence to suggest that the distribution of black individuals within the clusters varies significantly across the different temporal periods.

Now I will the conduct post-hoc test, Turkey Significant Difference (HSD) to help identify which specific clusters and temporal periods differ significantly in terms of the proportion of black individuals. I chose to add this step as it will give a more detailed understanding of the observed differences. 

```python
#importing the necessary libraries for this step

import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Flatten the multi-level index for the cluster_proportions dataframe
cluster_proportions_flat = cluster_proportions.copy()
cluster_proportions_flat.columns = ['_'.join(col).strip() for col in cluster_proportions.columns.values]
cluster_proportions_flat.reset_index(inplace=True)

#checking columns for cluster_proportions_flat: 
#print(cluster_proportions_flat.columns)

# Perform Tukey's HSD test
tukey_results = pairwise_tukeyhsd(
    endog=cluster_proportions_flat['P_r_o_p_o_r_t_i_o_n_ _B_l_a_c_k'],  # Use the correct column name
    groups=cluster_proportions_flat['Cluster'],
    alpha=0.05
)

# Showing the results
print(tukey_results.summary())
```

The results indicate statistically significant distinctions between Cluster 0 and Cluster 1, with Cluster 1 having a lower proportion of black individuals. However, there is no significant difference between Cluster 0 and Cluster 2. Notably, Cluster 2 exhibits a higher proportion of black individuals compared to Cluster 1. These findings contribute valuable insights into the specific variations in the racial composition of the clusters. This gives a nuanced understanding of the characteristics associated with each cluster.

In summary, the ANOVA test proportions of black individuals within clusters across different temporal periods. The F-statistic and p-value suggested a significant difference in these proportions. I will now explore the association between 'Perpetrator Race' and 'Temporal Period' globally. The chi-squared value and p-value indicated a significant association.
